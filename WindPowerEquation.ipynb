# The Wind Power Equation

*Kinetic Energy*: $\frac{1}{2}mv^2$

*Wind Power*: $\frac{1}{2}\rho \pi r^2 C_{p} CF v^3 NG NB$

## Wind Turbine Fundamentals

### Aerodynamics of wind turbines

Wind turbines extract kinetic engery from the wind and onvert it into mechanical and then electrical energy.
- **Betz's Law**: It states that the max achievable extraction of wind power by a turbine is 59.3% of the total kinetic energy in wind. This is known as the Betz Limit. 
$C_p = \frac{P_{turbine}}{P_{wind}}\leq 0.593$ *power coefficient*
- **Power Available in wind**: $P_{wind} = \frac{1}{2}\rho A V^3$
- **Swept Area**: $A = \pi R^2$ where R is the blade radius

### Blade element momentum theory

BEM combines Blade Element Theory and Momentum Theory to analyze the forces on the blades.
- **Momentum Theory**: Considers the turbine as an actuator disk that slows down the wind.
- **Blade Element Theory**: Divides the blade into small elements and analyzes the aerodynamic forces on each.
- **Axial Induction Factor (a)**: 

    $a = 1 - \frac{V_2}{V_1}$

    - $V_1$ is the free-stream wind speed.
    - $V_2$ is the wind speed at the rotor plane.

### Tip Speed Ratio ($\lambda$)

It is the ratio of the blade tip speed to the free-stream wind speed.

$\lambda = \frac{\omega R}{V}$

## Wind Farm Layout and Wake Effect

### Wake effect

When wind passes through a turbine, it loses kinetic energy, resulting in a wake (a region of reduced wind speed and increased turbulence downstream).

- Wake models:
    - Jensen's Park Model: A simple linear expansion model.
    
        $V(x) = V_0 (1 - \frac{2a}{(1 + 2k\frac{x}{D})^2})$

        - $V(x)$ is the wind speed at distance x downstream
        - $V_0$ is the free-stream wind speed
        - $a$ is the axial induction factor
        - $k$ is the wake decay constant (typically 0.075 offshore)
        - $D$ is the rotor diameter

### Wind Farm Layout Optimization

The goal is to position turbines to maximize energy production while minimizing wake losses
- Objective function: 

    $Max(E)= \sum{i=1}{N} P_i$

- Constraints:
    - Minimum distance: To avoid mechanical interference and excessive wake losses.

        $d_{ij} \geq d_{min}$
    - Land or Sea Area Boundaries: Turbines must be within the designated area.

## Wind Farm Control Techniques

### Yaw Control

Adjusting the turbine's orientation to align with wind direction.
- **Yaw Angle ($\gamma$)**: The angle between the turbine rotor plane and the wind direction.

### Pitch Control

Changing the blade pitch angle to regulate power output and loads
- **Blade Pitch Angle ($\theta$)**: The angle between the chord and the rotor plane.

### Torque control

Adjusting the generator torque to control the rotational speed.

### Wake Steering

Intentionally misaligning upstream turbines to deflect their wakes away from dowstream turbines.
- Objective: Reduce wake-induced power losses in dowstream turbines.

### Induction Control

Modifying the axial induction factor by adjusting the blade pitch or generator torque to affect the wake profile.

## Wind Turbine loads and structural analysis

### Aerodynamic loads

- Thrust Force: 

    $F_T = \frac{1}{2}\rho A V^2 C_T$

### Structural Dynamics

- Equation of Motion:

    $M\ddot{x} + C\dot{x} + Kx = F(t)$
    
    - M is the mass matrix
    - C is the damping matrix
    - K is the stiffness matrix
    - F(t) is the external force vector

### Fatigue Analysis

- **S-N curves**: Relate the stress amplitude (S) to the number of cycles to failure (N)
- **Miner's Rule**: $\sum{i}{}\frac{n_i}{N_i} = D$
    - $n_i$ is the number of cycles at stress level i.
    - $N_i$ is the number of cycles to failure at stress level i.
    - D i s the damage fraction (failure occurs when $D \geq 1$)
